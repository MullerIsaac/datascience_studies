# Projeto de Machine Learning de Previsão de Preço de Casas

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv('train.csv', index_col='Id')
df.head()
```

```python
df.dropna(axis=0, subset=["SalePrice"], inplace=True)
y = df.SalePrice
df.drop(['SalePrice'], axis=1, inplace=True)
```

```python
df_test = pd.read_csv("test.csv", index_col='Id')
```

```python
df.head()
```

```python
df.info()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(df, y, train_size=0.8, random_state=0) 
```

```python
# Selecionando as colunas categoricas
categorical = [columnname for columnname in df.columns if 
               df[columnname].nunique() < 10 and df[columnname].dtype == "object"]
```

```python
#Selecionando as colunas numéricas
numerical = [columnname for columnname in df.columns if 
               df[columnname].dtype in ["int64", "float64"]]
```

```python
columns = numerical + categorical
```

```python
X_train = X_train[columns].copy()
X_valid = X_valid[columns].copy()
X_test = df_test[columns].copy()
```

```python
X_train.head()
```

```python
# Preprocessamento das colunas numéricas
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Preprocessamento das colunas categoricas
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Pacote de transformações das colunas
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical),
        ('cat', categorical_transformer, categorical)
    ])

# Modelo
model = RandomForestRegressor(n_estimators=500, random_state=0)

# Pipeline de preprocessamento e criação de modelo
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('model', model)
                     ])

# Treinando o modelo
clf.fit(X_train, y_train)

# Predições
preds = clf.predict(X_valid)

print('MAE:', mean_absolute_error(y_valid, preds))
```

```python
preds_test = clf.predict(X_test)

output = pd.DataFrame({'Id': X_test.index,
                       'SalePrice': preds_test})
output.to_csv('submission.csv', index=False)
```
