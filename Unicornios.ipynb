# Análise de Dados de Startups Unicórnios

```python
#import de bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#carregamento dos dados
uni = pd.read_csv("unicorns till sep 2022.csv")
uni.shape
```

```python
uni.head()
```

```python
uni.columns
```

```python
uni["id"] = uni.index
```

```python
#Informações das colunas, é possível ver que não há valores nulos
uni.info()
```

```python
uni.nunique()
```

```python
uni.Industry.value_counts(normalize=True)
```

```python
per_setor = uni.groupby(["Industry"]).size()
per_setor = per_setor.sort_values()
per_setor
```

```python
plt.figure(figsize=(15,6))
plt.title("Setores da Indústria")
plt.bar(per_setor.index, uni.Industry.value_counts())
plt.xticks(rotation=90);
```

```python
per_pais = uni.Country.value_counts()
per_pais
```

```python
plt.figure(figsize=(15,6))
plt.title("10 maiores países que formam Unicórnios")
sns.barplot(x = per_pais.head(10), y = per_pais.index[0:10]);
```

```python
uni['Date Joined'] = pd.to_datetime(uni['Date Joined'])
```

```python
uni.dtypes
```

```python
uni['Month'] = pd.DatetimeIndex(uni['Date Joined']).month
uni['Year'] = pd.DatetimeIndex(uni['Date Joined']).year
```

```python
analise_country = uni.groupby(by= ["Country", "Month", "Year", "Company"]).count()['id'].reset_index()
```

```python
## Unicornios do Brasil
analise_country.loc[analise_country['Country'] == 'Brazil']
```

```python
uni['Valuation ($B)'] = pd.to_numeric(uni["Valuation ($B)"].apply(lambda row: row.replace('$', '')))
```

```python
uni.head()
```

```python
valuation_country = uni.groupby(['Country']).sum()['Valuation ($B)'].reset_index().sort_values('Valuation ($B)', ascending=False)
valuation_country
```

```python
plt.figure(figsize=(15,6))
plt.title("10 maiores países que formam Unicórnios")
sns.barplot(x= valuation_country['Country'], y = valuation_country['Valuation ($B)']);
plt.xticks(rotation=90);
```
